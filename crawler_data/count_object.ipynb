This notebook count all object datatype (ex: String, Party) on template library.

```python
import json
import re
```

```python
with open("./all_contract.json") as f:
    data = json.load(f)
```

```python
def is_obj_row(row, target):
    is_obj_row = (
        re.findall(f"\s{target}[\s|\n]", row) and
        "import" not in row and
        ("-->" in row or " o " in row)
    )
    return is_obj_row
```

```python
out = []
for key in data.keys():
    file_key = [s for s in data[key].keys() if "cto" in s][0]
    
    text = "".join(data[key]["/text/grammar.tem.md"])
    
    result = re.finditer("({[^{]*?)\w(?=\})}", text)

    targets = list(map(lambda x: x.string[x.start(): x.end()], result))
    #['{buyer}', '{amount}', '{seller}']
    targets = list(map(lambda x: re.sub("{|}", "", x), targets))
    #['buyer', 'amount', 'seller']
    
    has_match = False
    
    for row in data[key][file_key]:
        for target in targets:
            #'[exporter' -> 'exporter'
            target = re.sub("[^A-Za-z0-9\s]+", "", target)
            if is_obj_row(row, target):
                obj = re.findall(f"\s+(.*) {target}", row)
                #['--> Party']
                obj = obj[0].split(" ")[-1]
                #"Party"
                out.append(f"{target} => {obj}")
                has_match = True
    if has_match ==False:
        print("Not Found:", key)
        
```

```python
keywords = []
obj = []
for i in out:
    keywords.append(i.split(" => ")[0])
    obj.append(i.split(" => ")[1])
```

```python
import pandas as pd
df = pd.DataFrame()
```

```python
df["keyword"] = keywords
df["obj"] = obj
df["count"] = 1
```

```python
count = df.groupby("obj").sum()
```

```python
count = count.sort_values("count", ascending=False)
count
```
